```python
import time
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

#from selenium.webdriver.support.ui import WebDriverWait
#from selenium.webdriver.support import expected_conditions as EC
```

```python

data={'phone_numbers':[954302730, 977183791,965992236, 981827783, 993453473, 966504433, 934091769],"envio": 0}
cont=pd.DataFrame(data)
cont["msj"]="Prueba1"
cont.to_csv("contactos.csv", index=False)

df= pd.read_csv('contactos.csv')
df
```

```python
rng = np.random.default_rng()
delay = rng.integers(low=3, high=9, size=4).tolist()

# Initialize the web driver
options = webdriver.ChromeOptions() #carga configuración del webdriver
options.add_argument("--incognito")
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_argument("start-maximized")
options.add_argument('--no-sandbox')
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_experimental_option('useAutomationExtension', False)

driver=webdriver.Chrome('chromedriver',options=options)
driver.execute_script("Object.defineProperty(navigator, 'webdriver', {get: () => undefined})")

# Navigate to the Callbell login page
driver.get('https://dash.callbell.eu/users/sign_in')

# Wait for the page to load
time.sleep(delay[0])

# Find the input fields and enter the necessary login information
email_input = driver.find_element(By.ID,'user_email')
email_input.clear()
email_input.send_keys('gestionmatricula@gmail.com')
time.sleep(delay[1])

password_input = driver.find_element(By.ID, 'user_password')
password_input.clear()
password_input.send_keys('qwerty')
time.sleep(delay[2])
password_input.send_keys(Keys.RETURN)

# Wait for the dashboard page to load
time.sleep(rng.integers(low=8, high=12, size=1).tolist()[0])

# # Find the messaging icon and click it
messaging_icon = driver.find_element(By.XPATH,"//a[@href='/chat']").click()

time.sleep(delay[3])


for index, phone_number in df.phone_numbers.items():
    delay_loop = rng.integers(low=3, high=7, size=8).tolist()
    
    messaging_icon = driver.find_element(By.XPATH,"//div[@class='fullwidth-container chat-container']//div[@class='mr-4']").click()
    time.sleep(delay_loop[0])

    number_s = driver.find_element(By.ID, 'phone-form-control')
    number_s.clear()
    if index==0:
        number_s.send_keys('51'+str(phone_number))
    else:
        number_s.send_keys(str(phone_number))

    time.sleep(delay_loop[1])

    try:
        open_wi= driver.find_element(By.XPATH,"//div[@class='modal overflow-visible']//div[@class='modal_content-body']//span[@class='mr-2  text  ']")
    except:
        open_wi= driver.find_element(By.XPATH,"//div[@class='modal overflow-visible']//div[@class='modal_content-body']//div[@class='text-sm']")
    
    open_wi.click()
    time.sleep(delay_loop[2])

    hidden_element = driver.find_element(By.XPATH,"//div[@class='flex-grow h-screen main-content']/div[@class='fullwidth-container chat-container']/div[@data-react-class='v2/Chat']")
    driver.execute_script("arguments[0].style.visibility = 'visible';", hidden_element)
    
    time.sleep(delay_loop[3])
    text_m=hidden_element.find_element(By.XPATH,"//textarea[@placeholder='Type a message']")
    text_m.clear()
    time.sleep(delay_loop[4])

    text_m.send_keys('[Prueba 5] Le saludamos desde el Ministerio de Educación para recordarle que ya inició el registro de metas de atención para el año 2023\n')
    time.sleep(delay_loop[5])
    text_m.send_keys('Ponemos a su disposición el siguiente formulario para consultas relacionadas al registro de sus metas de atención https://docs.google.com/forms/d/e/1FAIpQLSe7vbRbsAbI9IA2VbioA9pwnUvegS5gEfxtA1NEOW6cZgCSYg/viewform \n')
    time.sleep(delay_loop[6])
    text_m.send_keys('Por favor, confirme con un "Sí", si recepcionó nuestro mensaje\n')

    time.sleep(delay_loop[7])
    df.loc[index, 'envio'] = 1

# Quit the web driver
driver.quit()
```
